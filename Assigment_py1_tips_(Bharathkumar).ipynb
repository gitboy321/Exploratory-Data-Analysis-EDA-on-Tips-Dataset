```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.load_dataset('tips')
```

```python
df = sns.load_dataset('tips')
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.barplot(x='sex', y='total_bill', data=df, hue="sex")
plt.title("Total Bill by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title("Total Bill Distribution by Gender")
plt.show()
```

```python
sns.barplot(x='sex', y='tip', data=df)
plt.title("Tips by Gender")
plt.show()
```

```python
sns.barplot(x='sex', y='tip', data=df, hue='day')
plt.title("Tips by Gender and Day")
plt.show()
```

```python
sns.countplot(x='sex',data=df, hue='day')
plt.title("Count of Male and Female by Day")
plt.show()
```

```python
sns.scatterplot(x='smoker', y='tip', data=df, hue='tip')
plt.title("Tips by Smoking Status")
plt.show()
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs Tip by Gender")
plt.show()
```

```python
sns.violinplot(x='sex', y='total_bill', data=df, hue='sex')
plt.title("Total Bill Distribution by Gender")
plt.show()
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(data=df, x="sex", y="tip")
plt.title("Average Tip by Gender")
plt.show()
```

```python
sns.barplot(x='day', y='tip', data=df, hue='day')
plt.title("Tips by Day")
plt.show()
```

```python
sns.countplot(x='smoker',data=df, hue='sex')
plt.title("Count of Smokers by Gender")
plt.show()
```
